```python
from dataclasses import dataclass

@dataclass
class Serie:
    titulo: str
    ano: str
    episodios: str
    classificacao: float
    pagina: str
    popularidade: str
    elenco_principal: list = None
    
```

```python


def infos_adicionais(pagina):
    driver.switch_to.new_window('tab')
    driver.get(pagina)
    driver.implicitly_wait(1)
    #scrolla um pouco pra baixo
    driver.execute_script("window.scrollTo(0, 2000);")
    print('taqui')
    
    popularidade = driver.find_element(By.XPATH, r'//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[3]/a/span/div/div[2]/div[1]').text
    print(f"Popularidade: {popularidade}")
    
    driver.implicitly_wait(1)
    
    tag_ul = driver.find_element(By.CSS_SELECTOR, "[data-testid='shoveler-items-container']")
    print("oi", tag_ul)
    lista_atores = tag_ul.find_elements(By.CSS_SELECTOR, "[data-testid='title-cast-item']")
    print(lista_atores)
    
    array_atores = []

    for atores in lista_atores:
        print(atores.find_element(By.CSS_SELECTOR, "[data-testid='title-cast-item__actor']").text)
        print(atores.find_element(By.CLASS_NAME, "sc-cd7dc4b7-4").text)
        array_atores.append(f"nome: {atores.find_element(By.CSS_SELECTOR, "[data-testid='title-cast-item__actor']").text}, personagem: {atores.find_element(By.CLASS_NAME, "sc-cd7dc4b7-4").text}")
        
    print(array_atores)
        
    driver.close()

    #Switch back to the old tab or window
    driver.switch_to.window(original_window)
        
    return popularidade, array_atores 
```

```python
import re

#biblioteca para express√µes regulares
def cria_serie(imdb_li_tag):
    titulo = imdb_li_tag.find_element(By.CLASS_NAME, "ipc-title__text").text.split(" ")[1:] #pegamos o texto do elemento e dividimos a string para pegar apenas o t√≠tulo
    titulo = " ".join(titulo) #juntamos a lista novamente em uma string
    print(titulo)
    ano = imdb_li_tag.find_element(By.XPATH, r'./div/div/div/div/div[2]/div[2]/span[1]').text.split("‚Äì")[0]
    episodios = imdb_li_tag.find_element(By.XPATH, r'./div/div/div/div/div[2]/div[2]/span[2]').text.split(" ")[0] #pegamos o texto do elemento e dividimos a string para pegar apenas o n√∫mero de epis√≥dios
    texto_classificacao = imdb_li_tag.find_element(By.CLASS_NAME, "ipc-rating-star").text #este texto est√° 'sujo', cont√©m o n√∫mero de votos tamb√©m
    pagina = imdb_li_tag.find_element(By.CLASS_NAME, "ipc-lockup-overlay").get_attribute("href")
    re_result = re.search(r'^\d\,\d',texto_classificacao) #aplicando a regex para extrair apenas o padr√£o 'digito ponto digito'
    
    ano = int(ano)
    episodios = int(episodios)
    
    classificacao = None 
    if re_result:
        classificacao = re.search(r'^\d\,\d',texto_classificacao).group(0) #string resultante da regex
        classificacao = float(classificacao.replace(",", ".")) # Aletera o separador decimal
    
    pupularidade, elenco_principal = infos_adicionais(pagina)
    
    print(titulo, ano, episodios, classificacao, pagina, pupularidade, elenco_principal)
    
    return Serie(
        titulo=titulo,
        ano=ano,
        episodios=episodios,
        classificacao=classificacao,
        pagina=pagina,
        popularidade=pupularidade,
        elenco_principal=elenco_principal
                 )
```

```python
lista_de_series = []
for series_tag in lista_atores:
        print("a")
        atores = cria_serie(series_tag)
        lista_de_series.append(atores)
```

```python
for i in range(10):
    print(lista_de_series[i])
```

```python
import json

with open("series.json", "w", encoding="utf-8") as arquivo:
    series_list = []
    for atores in lista_de_series:
        # Cria um dicion√°rio a partir do objeto Serie
        series_dict = {
            "titulo": atores.titulo,
            "ano": atores.ano,
            "episodios": atores.episodios,
            "classificacao": atores.classificacao,
            "pagina": atores.pagina,
            "popularidade": atores.popularidade,
            "elenco_principal": atores.elenco_principal
        }
        series_list.append(series_dict)
    
    # Salva a lista inteira de uma vez
    json.dump(series_list, arquivo, ensure_ascii=False, indent=4)
```
